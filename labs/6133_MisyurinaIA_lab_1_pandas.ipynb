Работа с пропущенными значениями:

удаление строк с пропущенными значениями;  
заполнение пропущенных значений средним значением по колонке.


```python
df.info()
```

```python
import random
import numpy as np

rows = df.shape[0]
types = df.dtypes
nullable_cols = [0, 1, 2, 3, 4, 5, 7, 8, 9]
cols = len(nullable_cols)

for i in range(500):
    row = df.iloc[random.randint(0, rows-1)].copy()
    row.iloc[nullable_cols[random.randint(0, cols-1)]] = np.nan
    df = pd.concat([df, row.to_frame().T], ignore_index = True)

df = df.astype(types)
```

```python
df.info()
```

```python
#удаление строк с пропущенными значениями
df.dropna(subset=["cut", "color", "clarity"], inplace=True)
```

```python
df.info()
```

```python
#заполнение пропущенных значений средним значением по колонке
numeric_cols = ['carat', 'depth', 'table', 'price', 'x', 'y', 'z']
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())
df = df.astype(types)
```

```python
df.info()
```

Создание нового поля, вычисленного на основе значений других полей:

через выражение на базе имеющихся колонок;  
через DataFrame.apply;  
через Series.apply.


```python
#Добавим price/carat
df["price/carat"] = (df["price"] / df["carat"]).round(2)
```

```python
#Добавим depth/table
df["depth/table"] = df.apply(lambda row: row["depth"] / row["table"], axis=1)
```

```python
#Добавим carat in gramms
df["carat in gramms"] = df["carat"].apply(lambda val: round(val * 0.2, 3))
```
